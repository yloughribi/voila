```python
import pandas as pd
import ipywidgets as widgets
from ipywidgets import interact
from ipywidgets import fixed
```

```python
covid = pd.read_csv("https://raw.githubusercontent.com/NHameleers/dtz2025-datasets/master/covid_2021.csv", parse_dates=['date'])
```

```python
# interact step 1: define a function
def select_country_data(iso_code, df):
  '''Takes iso_code and df as input. Returns the rows in
  df belonging to country with iso_code'''

  return df.loc[df['iso_code'] == iso_code, ]
```

```python
# interact step 2: pass that function and its parameters to interact
# this time with one parameter fixed, so no widget is created for df
print(f'The ISO codes you can choose are: {covid.iso_code.unique()}')
interact(select_country_data, iso_code='NLD', df=fixed(covid));
```

```python
interact(select_country_data, iso_code=['NLD', 'CAN', 'GBR', 'USA'], df=fixed(covid));
```

```python
# solution 5.1
import altair as alt

def visualize_country_data(iso_code, df):

  # 1. First we use the select_country_data function to select the subset
  # of df (here the covid data) with the specific iso_code
  data_to_viz = select_country_data(iso_code, df)

  # 2. use altair to create a chart (save it as variable called chart)
  # the chart should be: a line chart,
  # with 'yearmonth(date):T' on x-axis and
  # 'max(people_fully_vaccinated_per_hundred):Q' on y axis
  # showing location in the tooltip
  ### Your code here to create the chart
  chart = alt.Chart(data_to_viz).mark_line().encode(
      x='yearmonth(date):T',
      y='max(people_fully_vaccinated_per_hundred):Q',
      tooltip='location'
  )

  # display the chart object you created
  display(chart)

  # NB: most commonly a function would 'return' something.
  #  Yet, this gives an issue later on when using interactive_output.
  #  Therefore, we use display(chart) instead of return chart here.
```

```python
# solution 5.2
interact(visualize_country_data,
         iso_code=[('Netherlands', 'NLD'),
                                        ('Canada', 'CAN'),
                                        ('United Kingdom', 'GBR'),
                                        ('United States', 'USA')],
         df=fixed(covid));
```
